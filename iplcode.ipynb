```
df.drop(labels='mid', axis=1, inplace=True)
```

```
encoded_df = pd.get_dummies(data=df, columns=['venue','bat_team', 'bowl_team'])
```

```
encoded_df.head()
```

```
x_train = encoded_df.drop(labels='total', axis=1)[encoded_df['date'].dt.year <= 2016]
x_test = encoded_df.drop(labels='total', axis=1)[encoded_df['date'].dt.year >= 2017]

y_train = encoded_df[encoded_df['date'].dt.year <= 2016]['total']
y_test = encoded_df[encoded_df['date'].dt.year >= 2017]['total']

x_train.drop(labels='date', axis=True, inplace=True)
x_test.drop(labels='date', axis=True, inplace=True)
```

```
x_train.iloc[[0]].values[0]
```

```
x_train.head()
```

```
y_train.head()
```

**Linear Regression**

```
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train,y_train)
prediction=regressor.predict(x_test)
```

```
from sklearn import metrics
import numpy as np
print("Test Data")
print('MAE :', metrics.mean_absolute_error(y_test, prediction))
print('MSE:', metrics.mean_squared_error(y_test, prediction))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, prediction)))
prediction=regressor.predict(x_train)
print('-'*50)
print("Train Data")
print('MAE:', metrics.mean_absolute_error(y_train, prediction))
print('MSE:', metrics.mean_squared_error(y_train, prediction))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_train, prediction)))
```

**Ridge Regression with Hyperparameter Tuning**

```
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
ridge=Ridge()
parameters={'alpha':[1e-6,1e-4,1e-2,5,10,24,33,77,100,1000]}
clf1=GridSearchCV(ridge, parameters, cv=10, scoring='neg_mean_squared_error',return_train_score=True)
clf1.fit(x_train,y_train)
results1= pd.DataFrame.from_dict(clf1.cv_results_)
results1.head()
```

```
best_plist1 = clf1.best_params_
print("Best Hyperparameters are",best_plist1)
clf1.fit(x_train,y_train)
prediction=clf1.predict(x_test)
print("Test Data")
print('MAE :', metrics.mean_absolute_error(y_test, prediction))
print('MSE:', metrics.mean_squared_error(y_test, prediction))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, prediction)))
prediction=clf1.predict(x_train)
print('-'*50)
print("Train Data")
print('MAE:', metrics.mean_absolute_error(y_train, prediction))
print('MSE:', metrics.mean_squared_error(y_train, prediction))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_train, prediction)))
```

**Lasso Regression with Hyperparameter Tuning**

```
from sklearn.linear_model import Lasso
lasso=Lasso()
parameters={'alpha':[1e-2,5,10,24,33,77,100,1000]}
clf2=GridSearchCV(lasso, parameters, cv=10, scoring='neg_mean_squared_error',return_train_score=True)
clf2.fit(x_train,y_train)
results2= pd.DataFrame.from_dict(clf2.cv_results_)
results2.head()
```

```
best_plist2 = clf2.best_params_
print("Best Hyperparameters are",best_plist2)
clf2.fit(x_train,y_train)
prediction=clf2.predict(x_test)
print("Test Data")
print('MAE :', metrics.mean_absolute_error(y_test, prediction))
print('MSE:', metrics.mean_squared_error(y_test, prediction))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, prediction)))
prediction=clf2.predict(x_train)
print('-'*50)
print("Train Data")
print('MAE:', metrics.mean_absolute_error(y_train, prediction))
print('MSE:', metrics.mean_squared_error(y_train, prediction))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_train, prediction)))
```

**Decision Tree Regression with Hyperparameter Tuning**

```
from sklearn.ensemble import GradientBoostingRegressor
d1 = GradientBoostingRegressor()
params_grid = {'max_depth': [1, 5, 10],
               'min_samples_split':[5, 10]
               
               }
clf3 = GridSearchCV(d1, params_grid, cv=10, scoring='neg_mean_squared_error',return_train_score=True)
clf3.fit(x_train, y_train)
results3= pd.DataFrame.from_dict(clf3.cv_results_)
results3.head()
```

```
best_plist3 = clf3.best_params_
d1 = GradientBoostingRegressor(**best_plist3)
d1.fit(x_train,y_train)
prediction=clf2.predict(x_test)
print("Test Data")
print('MAE :', metrics.mean_absolute_error(y_test, prediction))
print('MSE:', metrics.mean_squared_error(y_test, prediction))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, prediction)))
prediction=d1.predict(x_train)
print('-'*50)
print("Train Data")
print('MAE:', metrics.mean_absolute_error(y_train, prediction))
print('MSE:', metrics.mean_squared_error(y_train, prediction))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_train, prediction)))
```

**Random Forest Regression with Hyperparameter Tuning**

```
from sklearn.ensemble import RandomForestClassifier
maxdepth=[2,5,10,20,30]
estimators = [10,50,100,150,200]
train_scores = []
test_scores = []
for i in estimators:
   for j in maxdepth:
      clf4 =RandomForestClassifier(max_depth=j,n_estimators=i)
      clf4.fit(x_train,y_train)
      train_sc = metrics.mean_absolute_error(y_train,clf4.predict(x_train))
      test_sc = metrics.mean_absolute_error(y_test,clf4.predict(x_test))
      print('Estimators',i,'Maxdepth',j,'Train Score',train_sc,'test Score',test_sc)
print("Done")
```

**GBDT Regression with Hyperparameter Tuning**

```
from xgboost import XGBClassifier
maxdepth=[2,5,10,15]
estimators = [5,10]
train_scores = []
test_scores = []
for i in estimators:
  for j in maxdepth:
    clf5 =XGBClassifier(default='gblinear',max_depth=j,n_estimators=i)
    clf5.fit(x_train,y_train)
    train_sc = metrics.mean_absolute_error(y_train,clf5.predict(x_train))
    test_sc = metrics.mean_absolute_error(y_test,clf5.predict(x_test))
    print('Estimators',i,'Maxdepth',j,'Train Score',train_sc,'test Score',test_sc)
print("Done")
```
